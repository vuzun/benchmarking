Each of the methods was run on a range of feature numbers (.py and .R files).

```python
#FPR/TPR calculation from the test set predictions and known class
#paras: known class list (binary); classifier predictions (float)
def get_fpr_tpr(testing_class, predicts):
    fpr, tpr, thres=roc_curve(testing_class, predicts)
    return fpr, tpr
```

```python
#visualising accuracy over varying feature size throug ROC curve
#params: list containing time, feature number, auc and test set predictions;
#  name of the model; whether testing class makes sense (not in PCA)
def rocs_over_feats(model_iterlist, name, testing_class_only=True):
    
    t_class=testing_class
    if not testing_class_only:
        t_class=np.concatenate([training_class,testing_class])
    
    rates=map(lambda x: get_fpr_tpr(t_class, x[3]), model_iterlist)


    plt.title('%s ROC curves for different feature size' % name)
    for i in range(len(rates)):
        if i==0:
            a_lab="20%% of the features - %0.2f AUC" % auc(rates[i][0], rates[i][1])
        elif i==(len(rates)-1):
            a_lab="All of the features - %0.2f AUC" % auc(rates[i][0], rates[i][1])
        plt.plot(rates[i][0], rates[i][1], color=colors[i].rgb,
                 label =  a_lab if i in [0,len(rates)-1] else None)
        plt.legend(loc = 'lower right')
        plt.plot([0, 1], [0, 1],color='grey', linestyle='--')
        plt.xlim([0, 1])
        plt.ylim([0, 1])
        plt.ylabel('True Positive Rate')
        plt.xlabel('False Positive Rate')
    plt.savefig("../graphs/"+name+"_feat_rocs.png")

from colour import Color
red = Color("grey")
colors = list(red.range_to(Color("red"),len(svm_results)))
```

```python
#extracting only time information
times_by_feat_by_alg=map(lambda l: map(lambda el: el[0], l),
                         [svm_results, gp_results, rf_results, pca_results, km_results])
#number of features
features=map(lambda x: x[1], svm_results)
```

There is a significant execution time between the classifiers and unsupervised methods. However, the difference could be due to language difference (R - Python). Still the rate of increase is at worst linear for all the methods and the classifiers used seem to be quite resistant to the increase in the number of features.

```python
plt.title("Model execution time across feature numbers")
model_colors=['r','g','b','y','black']
model_names=["SVM","GP","Random forest","PCA","k-means"]
model_lty=[":","--","-","-",":"]
for i in range(len(times_by_feat_by_alg)):
    plt.plot(features, times_by_feat_by_alg[i], model_colors[i], label=model_names[i], linestyle=model_lty[i])
    plt.legend(loc="upper left")
    plt.ylabel("Time (seconds)")
    plt.xlabel("Number of features")
plt.savefig("../graphs/times_v_features_all.png")#plt.figure()
```

Excluding the R methods, there are still noticeable, though small, differences between the classifiers.

Despite all the models being non-parametric, SVM is clearly slowed down by the number of features unlike random forests and Gaussian processes.

While there might be advantages to SVM, e.g. better support for kernel use to model nonlinearity better, random forests not performing worse on this type of problems (as can be seen by the AUC scores), their interpretability and speed make them a superior pick in this instance.
